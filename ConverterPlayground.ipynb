```python
import copy 

m1 = E.mesh(CLASS("walker/"),
            name="walker/sacrum",
            file="Geometry/femur.stl")
meshes = [m1,]

tendon = E.tendon()
actuator = E.actuator()
sensor = E.sensor()

asset = E.asset(
    E.texture(name="groundplane",
             type="2d",
             builtin="checker",
             rgb1="0.2 0.3 0.4",
             rgb2="0.1 0.2 0.3",
             mark="edge",
             markrgb="0.8 0.8 0.8",
             width="300",
             height="300"),
    E.material(CLASS("/"),
              name="groundplane",
              texture="groundplane",
              texrepeat="3. 3.",
              texuniform="true",
              reflectance="0.2"),
    *meshes,
)

def build_body_tree(root, elements, parents, children):
    chlds = []
    if root in children.keys():
        chlds = [build_body_tree(c, elements, parents, children) for c in children[root]]
    """        
      <joint armature="0" axis="1 0 0" damping="0" limited="false" name="walker/rootx" pos="0 0 0" stiffness="0" type="slide"/>
      <joint armature="0" axis="0 0 1" damping="0" limited="false" name="walker/rootz" pos="0 0 0" ref="1.25" stiffness="0" type="slide"/>
      <joint armature="0" axis="0 1 0" damping="0" limited="false" name="walker/rooty" pos="0 0 1.25" stiffness="0" type="hinge"/>
    """   
    joint_set = elements[root].find("Joint")
    joints = []
    if len(joint_set) > 0:
        if joint_set[0].tag == "PlanarJoint":
            jntx = E.joint(armature="0", axis="1 0 0", damping="0", limited="false", name="walker/rootx", pos="0 0 0", stiffness="0", type="slide")
            jnty = copy.deepcopy(jntx)
            jnty.attrib["axis"] = "0 0 1"
            jnty.attrib["name"] = "walker/rooty"
            jntr = copy.deepcopy(jntx)
            jntr.attrib["axis"] = "0 1 0"
            jntr.attrib["pos"] = "0 0 1.25"
            jntr.attrib["type"] = "hinge"
            jntr.attrib["name"] = "walker/rootr"
            joints.append(jntx)
            joints.append(jnty)
            joints.append(jntr)
    sites = []
    inertial = [E.inertial(pos="0. -0.17 0.", mass = "9.3014", fullinertia="0.1339 0.0351 0.1412 0. 0. 0.")]
    geom = [E.geom(CLASS("walker/"),
                   group="1",
                   name=("walker/%s" % root),
                   type="mesh",
                  mesh=("walker/%s" % root) )]
    return E.body(*joints,
                  *inertial,
                  *chlds,
                  *sites,
                  *geom,
                  name="walker/%s" % root)

worldbody = E.worldbody(
    E.geom(CLASS("/"), name="groundplane", type="plane", size='8. 8. 0.5', material="groundplane"),
    E.camera(CLASS("/"), name="top_camera", fovy="10.0581471636", pos="0. 0. 100.", zaxis="0. 0. 1."),
    E.body(
        E.joint(),
        E.joint(),
        E.joint(),
        E.light(),
        build_body_tree("ground", elements, parents, children),
        name="walker/")
)

mujoco_doc = E.mujoco(
    E.compiler(boundmass = "1e-05",
              boundinertia="1e-11",
              balanceinertia="true",
              coordinate="local",
              angle="radian",
              eulerseq="xyz"),
    E.option(timestep="0.002",
            cone="elliptic",
            noslip_iterations="5",
            noslip_tolerance="0.0"),
    E.size(njmax="500",
          nconmax="150"),
    E.visual(
        E.headlight(ambient="0.4 0.4 0.4",
                   diffuse="0.8 0.8 0.8",
                   specular="0.1 0.1 0.1"),
        E.map(znear="0.01"),
        E.scale(forcewidth="0.01",
               contactwidth="0.06",
               contactheight="0.01",
               jointwidth="0.01",
               framelength="0.3",
               framewidth="0.01"),
    ),
    E.default(
        E.default(CLASS("/") ),
        E.default(
            E.joint(solimplimit="0. 0.99 0.01",
                    stiffness="1.0",
                    armature="0.01",
                    damping="0.05"),
            E.equality(solimp="0 0.999 0.0001"),
            E.tendon(stiffness="40.0"),
            E.general(ctrllimited="true",
                      ctrlrange="-1. 0.",
                      dyntype="filter",
                      dynprm="0.1"),
            E.default(CLASS("walker/wrapping"),
                     E.geom(type="cylinder",
                            contype="0",
                            conaffinity="0",
                            group="3",
                            rgba="0.5 0.5 0.5 0.2",
                            mass="0.0")),
            E.default(CLASS("walker/upper_joint"),
                     E.joint(stiffness="5.0",
                            armature="0.1",
                            damping="2.0")),
            E.default(CLASS("walker/direct"),
                     E.general(ctrlrange="-1. 1.")),
            CLASS("walker/"))
    ),
    asset,
    worldbody,
    tendon,
    actuator,
    sensor,
    model="floor")
```

```python
print(etree.tostring(mujoco_doc, pretty_print=True).__str__().replace("\\n","\n").replace("\\t","\t"))

tree = etree.ElementTree(mujoco_doc)
tree.write(file_out, pretty_print=True, xml_declaration=True,   encoding="utf-8")
```
